# 用途別 matplotlib　グラフ作成用コード

## csvをpandasで読み込んでグラフ表示
- fontsizeの一括指定
    - plt.rcParams["font.size"]

- 見切れてしまう際の表示調節
    - plt.rcParams["figure.subplot.bottom"] = 0.15
    - plt.rcParams["figure.subplot.left"] = 0.18
- 凡例の指定
    - plt.legend(bbox_to_anchor=(1, 1), loc="upper right", borderaxespad=0)
    - [ドキュメント](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html)
    
- log scale 指定
    - plt.xscale("log"): 


```python
import pandas as pd
from matplotlib import pyplot as plt

fs = 96000
filepath = "../../data/attribute_graphs/"
filename = "sensordata.csv"
df = pd.read_csv(filepath + filename)

# df.columns = ["acc_x", "acc_y", "acc_z", "gyro_x", "gyro_y", "gyro_z"] # 列名を新たに指定

t = df["t"] / fs # サンプル番号 -> サンプルを取得した秒数に変換

plt.rcParams["font.size"] = 12 # グラフのフォントサイズを一括指定


# 凡例名の指定
plt.plot(t, df["acc_x"], label="acc_x")


# ラベル指定
plt.xlabel("time [s]")
plt.ylabel("Attribute values")

# 範囲指定
#plt.xlim([25, 50])
#plt.ylim([-1, 1])

# logスケール指定
#plt.xscale("log")
#plt.yscale("log")

# ラベルやグラフが見切れてしまう際の調節
#plt.rcParams["figure.subplot.bottom"] = 0.15
#plt.rcParams["figure.subplot.left"] = 0.18


# 凡例の指定
plt.legend(bbox_to_anchor=(1, 1), loc="upper right", borderaxespad=0)

plt.savefig("../../data/attribute_graphs/attribute-graph-statistics-frequency.png")
```

## 複数グラフの表示 (axsが一次元)
- [Document](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html)

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import glob
import os



#files = os.listdir(dataFilepath)
#files = ['3000.csv']
filepath = '../../data/baseexperiment/'
files = glob.glob(filepath+'*.csv')
print(files)

for file in files:
    try:
        print('File:', file)
        file = os.path.splitext(os.path.basename(file))[0]

        csv_file = pd.read_csv(filepath+file+'.csv')

        csv_file.columns = ["acc_x", "acc_y", "acc_z", "gyro_x", "gyro_y", "gyro_z"]

#リスト形式
#print(csv_file)
        print(csv_file.shape)

        t = range(csv_file.shape[0])
        
        

        fig, axs = plt.subplots(nrows=3, figsize=(10,10))   # 縦並び
        # fig, axs = plt.subplots(ncols=3, figsize=(10,10)) # 横並び

        axs[0].plot(t, csv_file['acc_x'], linewidth=2)
        axs[0].set_title('acc_x')
        axs[0].set_xlabel('t')
        axs[0].set_ylabel('acc_x raw')
        #axs[0].set_xlim(0, DATASIZE)
        #axs[0].set_ylim(-0.50, -0.20)
        #axs[0].grid(True)

        axs[1].plot(t, csv_file['acc_y'], linewidth=2)
        axs[1].set_title('acc_y')
        axs[1].set_xlabel('t')
        axs[1].set_ylabel('acc_y raw')
        #axs[1].set_xlim(0, DATASIZE)
        #axs[1].set_ylim(0.11, 0.16)
        #axs[1].grid(True)

        axs[2].plot(t, csv_file['acc_z'], linewidth=2)
        axs[2].set_title('acc_z')
        axs[2].set_xlabel('t')
        axs[2].set_ylabel('acc_z raw')
        #axs[2].set_xlim(0, DATASIZE)
        #axs[2].set_ylim(0.72, 0.79)
        #axs[2].grid(True)

        fig.suptitle(file+'.csv')
        #fig.show()
        plt.savefig(filepath + 'figs/'+file+ 'acc' +'.png')
    except Exception as other:
        print(file, other)

```
